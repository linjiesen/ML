```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny,linestyle="-.")
plt.plot(x,cosy,color="red",linestyle="--")  
plt.xlim(-5,15)                              # Adjust the limit of x
plt.ylim(0,1.5)                              # Adjust the limit of y
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.axis([-1,11,-2,2])                        # x[-1,11]  y[-2,2]
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.xlabel("x axis")                  # what will display on x
plt.ylabel("y value")                 # what will diaplay on y

plt.show()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",label="cos(x)",linestyle="--")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y value")
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",label="cos(x)",linestyle="--")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("Welcome to The ML World!")
```
